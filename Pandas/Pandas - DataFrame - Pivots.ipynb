```python
import pandas as pd    #python data analysis

#format settings
pd.options.display.float_format = '{:.4f}'.format

#read from CSV
str_path_data = "..//..//data//"
dt_world_raw = pd.read_csv(str_path_data + "worldstats.csv")
dt_world_raw.head(10)
```

### Requires Index Columns for Pivot

```python

#reading data with INDEX : Used for stacking / unstacking
dt_world = pd.read_csv(str_path_data + "worldstats.csv",index_col=["country","year"])
dt_world.head(10)
```

```python
s_dt_world = dt_world.stack()
s_dt_world.head(10)
```

```python
s_dt_world.unstack(level = [0]).head(10) # level is stacking on column index
```

```python
s_dt_world.unstack(level = [1]).head(10)  # level is stacking on column index
```

```python
#Filter then Index then stack

#Filter performs the WHERE condition
#SetIndex prepares the Levels for the Stacking
#Unstacking group by the target level.
dt_world_raw [ dt_world_raw["year"].isin(["2007","2008"]) ].set_index(["country","year"]).unstack(level = [0])
```

```python
#Filter then Index then stack

#Filter performs the WHERE condition
#SetIndex prepares the Levels for the Stacking
#Unstacking group by the target level.
dt_world_raw [ dt_world_raw["year"].isin(["2007","2008"]) ].set_index(["country","year"]).unstack(level = "country")
```
