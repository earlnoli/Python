```python
import pandas as pd
import numpy as np

#read from CSV
#
data = pd.read_csv("train.csv")
data[:10]
```

```python
from sklearn.preprocessing import Imputer

data2 = data[['PassengerId', 'Survived', 'Pclass','Age','SibSp','Parch','Fare']].copy()

data2[:20]
```

```python
fill_NaN = Imputer(missing_values=np.nan, strategy='median', axis=0)

#columns_names are gone after imputation, so we need to bring them back using columns = data2.columns
#data transformation makes the data types convert to floats.  use .astype($original_data$.dtypes.to_dict())
data2_im = pd.DataFrame(fill_NaN.fit_transform(data2), columns = data2.columns).astype(data2.dtypes.to_dict())


data2_im[:20]
```
